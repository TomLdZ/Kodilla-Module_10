```python
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats
```

```python
import random  

nums_100 = []  
mu = 3
sigma = 1
    
for i in range(100):  
    temp = random.gauss(mu, sigma)  
    nums_100.append(temp)  

nums_10k = []  
mu = 3
sigma = 1
    
for i in range(10000):  
    temp = random.gauss(mu, sigma)  
    nums_10k.append(temp)

fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(12, 5))
axes[0].hist(nums_100, bins = 20)  
axes[1].hist(nums_10k, bins = 20)
plt.show() 
```

## Średnia

```python
mean_100 = np.round(np.mean(nums_100), decimals=2)
mean_10k = np.round(np.mean(nums_10k), decimals=2)
print(f'Mean 100 = {mean_100}, Mean 10k = {mean_10k}')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(12, 5))
axes[0].hist(nums_100, bins = 20)
axes[0].axvline(x = mean_100, color='red', label = 'Średnia')
axes[0].legend(loc='upper right')
axes[1].hist(nums_10k, bins = 20)
axes[1].axvline(x = mean_10k, color='red', label = 'Średnia')
axes[1].legend(loc='upper right')
plt.show() 
```

## Mediana

```python
median_100 = np.round(np.median(nums_100), decimals=2)
median_10k = np.round(np.median(nums_10k), decimals=2)
print(f'Median 100 = {median_100}, Median 10k = {median_10k}')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(12, 5))
axes[0].hist(nums_100, bins = 20)
axes[0].axvline(x = mean_100, color='red', label = 'Średnia')
axes[0].axvline(x = median_100, color='green', label = 'Mediana')
axes[0].legend(loc='upper right')
axes[1].hist(nums_10k, bins = 20)
axes[1].axvline(x = mean_10k, color='red', label = 'Średnia')
axes[1].axvline(x = median_100, color='green', label = 'Mediana')
axes[1].legend(loc='upper right')
plt.show() 
```

## Moda

```python
mode_100 = stats.mode(nums_100, keepdims=True)
mode_10k = stats.mode(nums_10k, keepdims=True)
print(f'Mode 100 = {mode_100}, Mode 10k = {mode_10k}')
```

Każda wartość występuje tylko raz i funkcja wybrała najmniejsze wartości ze zbiorów.

## Kwartyle

```python
q0_100 = np.round(np.quantile(nums_100, 0.0), decimals=2)
q1_100 = np.round(np.quantile(nums_100, 0.25), decimals=2)
q3_100 = np.round(np.quantile(nums_100, 0.75), decimals=2)
q4_100 = np.round(np.quantile(nums_100, 1.0), decimals=2)
q0_10k = np.round(np.quantile(nums_10k, 0.0), decimals=2)
q1_10k = np.round(np.quantile(nums_10k, 0.25), decimals=2)
q3_10k = np.round(np.quantile(nums_10k, 0.75), decimals=2)
q4_10k = np.round(np.quantile(nums_10k, 1.0), decimals=2)
print(f'Q0 100 = {q0_100}, Q0 10k = {q0_10k}')
print(f'Q1 100 = {q1_100}, Q1 10k = {q1_10k}')
print(f'Q2 100 = {median_100}, Q2 10k = {median_10k}')
print(f'Q3 100 = {q3_100}, Q3 10k = {q3_10k}')
print(f'Q4 100 = {q4_100}, Q4 10k = {q4_10k}')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(12, 5))
axes[0].hist(nums_100, bins = 20)
axes[0].axvline(x = mean_100, color='red', label = 'Średnia')
axes[0].axvline(x = median_100, color='green', label = 'Mediana')
axes[0].axvline(x = q0_100, color='black', label = 'Min')
axes[0].axvline(x = q1_100, color='black', label = 'Q1')
axes[0].axvline(x = q3_100, color='black', label = 'Q3')
axes[0].axvline(x = q4_100, color='black', label = 'Max')
axes[0].legend(loc='upper right')
axes[1].hist(nums_10k, bins = 20)
axes[1].axvline(x = mean_10k, color='red', label = 'Średnia')
axes[1].axvline(x = median_100, color='green', label = 'Mediana')
axes[1].axvline(x = q0_10k, color='black', label = 'Min')
axes[1].axvline(x = q1_10k, color='black', label = 'Q1')
axes[1].axvline(x = q3_10k, color='black', label = 'Q3')
axes[1].axvline(x = q4_10k, color='black', label = 'Max')
axes[1].legend(loc='upper right')
plt.show() 
```

## Zakres

```python
range_100 = q4_100 - q0_100
range_10k = q4_10k - q0_10k
print(f'Range 100 = {range_100}, Range 10k = {range_10k}')
```

## IQR

```python
iqr_100 = q3_100 - q1_100
iqr_10k = q3_10k - q1_10k
print(f'IQR 100 = {iqr_100}, IQR 10k = {iqr_10k}')
```

## BoxPlot

```python
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(12, 5))
axes[0].boxplot(nums_100)
axes[1].boxplot(nums_10k)
plt.show()
```

## Wariancja

```python
variance_100 = np.var(nums_100, ddof=1)
variance_10k = np.var(nums_10k, ddof=1)
print(f'Variance 100 = {variance_100}, Variance 10k = {variance_10k}')
```

## Odchylenie standardowe

```python
standard_deviation_100 = np.std(nums_100, ddof=1)
standard_deviation_10k = np.std(nums_10k, ddof=1)
print(f'Standard deviation 100 = {standard_deviation_100}, Standard deviation 10k = {standard_deviation_10k}')
```
