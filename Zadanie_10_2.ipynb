```python
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats
```

```python
def gauss(x, sigma=1, mu=0):
    return stats.norm.pdf(x, mu, sigma)
x = np.linspace(-10, 10, 100)
y =  gauss(x, sigma=1, mu=3)
plt.figure(figsize=(12, 8))
plt.plot(x, y, label='odchylenie standardowe = 1, średnia = 3')
plt.legend(loc='upper left')
plt.show()
```

## Średnia

```python
mean_y = np.mean(y)
print(mean_y)
```

```python
plt.hist(y, bins=20)
plt.axvline(x = mean_y, color='red', label = 'Średnia')
plt.legend(loc='upper right')
plt.show()
```

## Mediana

```python
median_y = np.median(y)
print(median_y)
```

```python
plt.hist(y, bins=20)
plt.axvline(x = mean_y, color='red', label = 'Średnia')
plt.axvline(x = median_y, color='green', label = 'Mediana')
plt.legend(loc='upper right')
plt.show()
```

## Moda

```python
mode_y = stats.mode(y, keepdims=True)
print(mode_y)
```

```python
plt.hist(y, bins=20)
plt.axvline(x = mean_y, color='red', label = 'Średnia')
plt.axvline(x = median_y, color='green', label = 'Mediana')
plt.axvline(x = mode_y[0], color='orange', label = 'Moda')
plt.legend(loc='upper right')
plt.show()
```

## Kwartyle

```python
q0 = np.quantile(y, 0.0)
q1 = np.quantile(y, 0.25)
q3 = np.quantile(y, 0.75)
q4 = np.quantile(y, 1.0)
print(f'Q0: {q0}')
print(f'Q1: {q1}')
print(f'Q2: {median_y}')
print(f'Q3: {q3}')
print(f'Q4: {q4}')
```

```python
plt.hist(y, bins=20)
plt.axvline(x = mean_y, color='red', label = 'Średnia')
plt.axvline(x = median_y, color='green', label = 'Mediana')
plt.axvline(x = mode_y[0], color='orange', label = 'Moda')
plt.axvline(x = q0, color='black', label = 'Min')
plt.axvline(x = q1, color='black', label = 'Q1')
plt.axvline(x = q3, color='black', label = 'Q3')
plt.axvline(x = q4, color='black', label = 'Max')
plt.legend(loc='upper right')
plt.show()
```

## Zakres

```python
range_y = max(y) - min(y)
print(range_y)
```

## IQR

```python
iqr = q3 - q1
print(iqr)
```

## BoxPlot

```python
plt.boxplot(y)
plt.show()
```
