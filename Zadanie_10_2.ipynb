```python
# manipulacja danymi
import numpy as np
import pandas as pd

# wizualizacja danych
import matplotlib.pyplot as plt
import seaborn as sns

# statystyczna analiza danych
from scipy import stats
```

```python
import random  

nums_100 = []  
mu = 3
sigma = 1
    
for i in range(100):  
    temp = random.gauss(mu, sigma)  
    nums_100.append(temp)  

nums_10k = []  
mu = 3
sigma = 1
    
for i in range(10000):  
    temp = random.gauss(mu, sigma)  
    nums_10k.append(temp)

fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(10, 5))
axes[0].hist(nums_100, bins = 20)  
axes[1].hist(nums_10k, bins = 20)
plt.show() 
```
